## シンプルなRNNの問題点
前回作ったRNNはループする構造を持ち、過去の情報を記憶することができた。しかし、あまり性能はよくない。

そこでRNNに変わってLSTMやGRUといったモデルが登場してきた。通常はこれらをRNNと呼び、前回のRNNは明示的に「シンプルなRNN」や「エルマン」等と呼ばれる。

LSTMやGRUにはゲートと呼ばれる仕組みが備わっている。まずは詳しくシンプルなRNNの問題点を見た後、ゲートの仕組みを解説する。

簡単にRNNを復習すると、入力$x_t$と前回の出力$h_{t-1}$から$h_t$を出力する。

$$
h_t = \tanh(h_{t-1}W_h + x_tW_x + b)
$$

### 問題点①tanhによる勾配消失

時刻T分のバッチに対して損失を計算するとき、上記計算の逆をT回行うことになる。

足し算は単に値を流すだけだが、tanhとmatmul演算についてはどうだろうか。まずはtanhについてみていく



```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(-5, 5, 0.1)
y = np.array([math.tanh(j) for j in x])
dy = 1-y**2
plt.plot(x, y)
plt.plot(x, dy)
plt.show()
```

青色がtanhでオレンジがその導関数になる。

すなわち、導関数は最大値が1なので、この導関数を通るたびに勾配はどんどん小さくなり、初めの入力に近いRNNレイヤでは勾配が消失してしまう。この原因によってシンプルなRNNでは長期間の記憶ができていない。

### 問題点②matmulレイヤによる勾配消失・爆発

次にmatmulレイヤについてみていく。

ここでは、matmulレイヤの処理だけについて注目してみる。

損失から流れてきた値$dh$はmatmulを通るたびに$Wh$の転置行列をかける。すなわち、時刻T後の損失が初めのRNNにたどり着くころには勾配は次のようになっている。

$$
dh_0 = dh(W_h^t)^T
$$

計算後の$dh$はざっくりと計算すれば、行列$W_h$の最大の固有値が1より大きければ非常に大きな値つまり勾配爆発し、1より小さければ勾配消失となってしまう。

```python
N = 2 #ミニバッチ数
H = 3 # 隠れ状態ベクトルの次元
T = 20 #時系列データの長さ

dh = np.ones((N, H))
np.random.seed(5)
Wh = np.random.randn(H, H)

norm_list = []
for t in range(T):
    dh = np.dot(dh, Wh.T)
    norm = np.sqrt(np.sum(dh**2)) / N
    norm_list.append(norm)

plt.plot([t for t in range(T)], norm_list)
plt.show()
```

```python
np.linalg.eig(Wh)
```

```python
N = 2 #ミニバッチ数
H = 3 # 隠れ状態ベクトルの次元
T = 20 #時系列データの長さ

dh = np.ones((N, H))
np.random.seed(5)
Wh = np.random.randn(H, H) * 0.5 # 1より小さくしてみる
norm_list = []
for t in range(T):
    dh = np.dot(dh, Wh.T)
    norm = np.sqrt(np.sum(dh**2)) / N
    norm_list.append(norm)

plt.plot([t for t in range(T)], norm_list)
plt.show()
```

## 勾配爆発への解決策
勾配爆発には定番の解決策があり、勾配クリッピングと呼ばれる。

勾配クリッピングでは、ネットワーク内全てのパラメータに対する勾配を結合した行列$g$に対して次のような操作をする

$$
if (\|g\| >= threshold)\\
   g = \frac{threshold}{\|g\|}g
$$

thresoldは閾値で、この値を超えたときにパラメータ全体を更新する。

## 勾配消失への解決策
勾配消失を解決するには根本的にモデルを改良する必要があり、その代表的なものにLSTMやGRUがある。

LSTMはRNNと比べると次のような違いがある。

<img src="lstm2.png">

RNNは前回からの出力と今回入力を受け取ったが、LSTMでは新たにc（記憶セル）と前回出力、今回入力を受け取って、h,c二つの値を出力する。

h(隠れ状態)はネットワーク内を流れていくが、c(記憶セル)はLSTM内でのみ流れていく。

LSTMの中身をもう少し詳しく描くと次のようになる。

<img src="lstm_detail.png">

計算だけを抜き出して書くと次のようになる

$$
f = \sigma(x_tW_x^f + h_{t-1}^f + b^f)\\
g = \tanh(x_tW_x^g + h_{t-1}^g + b^g)\\
i = \sigma(x_tW_x^i + h_{t-1}^i + b^i)\\
o = \sigma(x_tW_x^o + h_{t-1}^o + b^o)\\
c_t = f \odot c_{t-1} + g \odot i\\
h_t = o\odot \tanh(c_t)
$$



$\sigma$はシグモイド関数を表している。シグモイド関数はある値を確率へ変換するという性質がある。このシグモイド関数が入力値の何%を次へ流すかというゲートの役割を担っている。ゲートの開き具合はそれぞれの重み関数、バイアスと二つの入力によって決まる。

まず、$o$はoutputゲートを表し、LSTMの出力をどれくらい次へ流すかを決める。

$f$はforgetゲートを表し、記憶セルに対して不要な記憶を忘れさせる。

forgetゲートにより記憶セルは情報を削除される。このままでは忘れる一方なので$g$によって新しく記憶が追加される。

ただ単に新しい情報を与えるだけでなく、$i$inputゲートによりどの程度記憶するのかを決める。

つまり前回の記憶セルに対してある割合で情報を忘れ、新しい情報のある割合部分を記憶し、次へ記憶として流す。

逆に、忘れるべき情報に対する勾配は消失し、重要な情報と判断された場合は前のセルへ勾配を消失することなく流すことができる。


LSTMは基本的にはこのような構造になっている。

LSTMと勾配クリッピングを使うことで、RNNの二つの問題点に対応することができた。
